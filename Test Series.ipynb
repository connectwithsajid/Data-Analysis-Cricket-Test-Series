```python
for i in range(len(teams)):
    Total_Matches_Played = 0
    Total_Wins = 0
    Total_Lose = 0
    Total_Draw = 0
    
    for j in range(len(dataset)):
        if(dataset.at[j,'Team'] == teams[i] or dataset.at[j,'Opponent'] == teams[i]):
            Total_Matches_Played += 1
        if(dataset.at[j,'Team'] == teams[i] and dataset.at[j,'Result'] == "Won"):
            Total_Wins += 1
            
        if(dataset.at[j,'Team'] == teams[i] and dataset.at[j,'Result'] == "Lost"):
            Total_Lose += 1
            
        if(dataset.at[j,'Team'] == teams[i] and dataset.at[j,'Result'] == "Drawn"):
            Total_Draw += 1
        
        if(dataset.at[j,'Opponent'] == teams[i] and dataset.at[j,'Result'] == "Won"):
            Total_Lose += 1
            
        if(dataset.at[j,'Opponent'] == teams[i] and dataset.at[j,'Result'] == "Lost"):
            Total_Wins += 1
            
        if(dataset.at[j,'Opponent'] == teams[i] and dataset.at[j,'Result'] == "Drawn"):
            Total_Draw += 1
    
    print("===================================")
    print("Team : " + str(teams[i]))
    print("Total Matches Played :" + str(Total_Matches_Played))
    print("Total Matches Won :" + str(Total_Wins))
    print("Total Matches Lost :" + str(Total_Lose))
    print("Total Matches Drawn :" + str(Total_Draw))
    
    Teams_Report.append({'Team' : teams[i] , 'Total Matches Played':Total_Matches_Played , 'Wins':Total_Wins , 'Lost':Total_Lose , 'Drawn':Total_Draw})
```

```python
Team1 = []
Team2 = []
Team1Wins = []
Team2Wins = []
Team1Lost = []
Team2Lost = []
Team1Drawn = []
Team2Drawn = []
for team1 in teams:
    for team2 in teams:
        team1Wins = 0
        team2Wins = 0
        team1Lost = 0
        team2Lost = 0
        team1Drawn = 0
        team2Drawn = 0
        if(team1 != team2):
            for i in range(len(dataset)):
                if(dataset.at[i,'Team'] == team1 and dataset.at[i,'Opponent'] == team2):
                    if(dataset.at[i,'Result'] == 'Won'):
                        team1Wins += 1
                        team2Lost += 1
                        
                    if(dataset.at[i,'Result'] == 'Lost'):
                        team1Lost += 1
                        team2Wins += 1
                    
                    if(dataset.at[i,'Result'] == 'Drawn'):
                        team1Drawn += 1
                        team2Drawn += 1
                
                if(dataset.at[i,'Team'] == team2 and dataset.at[i,'Opponent'] == team1):
                    if(dataset.at[i,'Result'] == 'Won'):
                        team2Wins += 1
                        team1Lost += 1
                        
                    if(dataset.at[i,'Result'] == 'Lost'):
                        team2Lost += 1
                        team1Wins += 1
                    
                    if(dataset.at[i,'Result'] == 'Drawn'):
                        team2Drawn += 1
                        team1Drawn += 1
            
            Team1.append(team1)
            Team2.append(team2)
            Team1Wins.append(team1Wins)
            Team2Wins.append(team2Wins)
            Team1Lost.append(team1Lost)
            Team2Lost.append(team2Lost)
            Team1Drawn.append(team1Drawn)
            Team2Drawn.append(team2Drawn)
            
            
```

```python
data = {'Team1':Team1,'Team2':Team2,'Team1Wins':Team1Wins,'Team2Wins':Team2Wins,'Team1Lost':Team1Lost,'Team2Lost':Team2Lost
        ,'Team1Drawn':Team1Drawn,'Team2Drawn':Team2Drawn}
Processed_Data = pd.DataFrame(data)
```

```python
for i in range(len(Processed_Data)):
    t1 = Processed_Data.at[i,'Team1']
    t2 = Processed_Data.at[i,'Team2']
    for i in range(len(Processed_Data)):
        if(t1 == Processed_Data.at[i,'Team2'] and t2 == Processed_Data.at[i,'Team1']):
            Processed_Data.loc[i,'Team1'] = np.nan
            Processed_Data.loc[i,'Team2'] = np.nan
```
